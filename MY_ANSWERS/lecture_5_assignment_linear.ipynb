```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Step 1: Load the dataset
data = pd.read_excel("datasets/ENB2012_data.xlsx")

# Step 2: Data Exploration and Preprocessing (if necessary)
# Check for missing values
print(data.info())

# Step 3: Feature Selection
# Select features X1 to X8
X = data[['X1', 'X2', 'X3', 'X4', 'X5', 'X6', 'X7', 'X8']]
y1 = data['Y1']  # Target variable for Heating Load
y2 = data['Y2']  # Target variable for Cooling Load

# Step 4: Split the Data
X_train, X_test, y1_train, y1_test = train_test_split(X, y1, test_size=0.2, random_state=42)
X_train, X_test, y2_train, y2_test = train_test_split(X, y2, test_size=0.2, random_state=42)

# Step 5: Choose a Regression Model
model1 = LinearRegression()
model2 = LinearRegression()

# Step 6: Train the Model
model1.fit(X_train, y1_train)
model2.fit(X_train, y2_train)

# Step 7: Evaluate the Model
y1_pred = model1.predict(X_test)
y2_pred = model2.predict(X_test)

print("Heating Load:")
print("Mean Squared Error:", mean_squared_error(y1_test, y1_pred))
print("R-squared:", r2_score(y1_test, y1_pred))

print("\nCooling Load:")
print("Mean Squared Error:", mean_squared_error(y2_test, y2_pred))
print("R-squared:", r2_score(y2_test, y2_pred))

# Step 8: Visualize the Results
plt.scatter(y1_test, y1_pred)
plt.xlabel("Actual Heating Load")
plt.ylabel("Predicted Heating Load")
plt.title("Actual vs. Predicted Heating Load")
plt.show()

plt.scatter(y2_test, y2_pred)
plt.xlabel("Actual Cooling Load")
plt.ylabel("Predicted Cooling Load")
plt.title("Actual vs. Predicted Cooling Load")
plt.show()
```
